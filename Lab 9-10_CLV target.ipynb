# Processing Data

## 1. Dealing with outliers

```python
#eliminating outliars from tca 
iqr = np.percentile(df["tca"], 75) - np.percentile(df["tca"], 25)
upper_limit = np.percentile(df["tca"], 75) + 1.5 * iqr
lower_limit = np.percentile(df["tca"], 25) - 1.5 * iqr

mask_not_outlier = (df.tca <= upper_limit) & (df.tca >= lower_limit)
df_clean_tca = df[mask_not_outlier]
display(df_clean_tca)
sns.boxplot(df_clean_tca['tca'])
```

```python
#eliminating outliars from clv 
iqr = np.percentile(df_clean_tca["clv"], 75) - np.percentile(df_clean_tca["clv"], 25)
upper_limit = np.percentile(df_clean_tca["clv"], 75) + 1.5 * iqr
lower_limit = np.percentile(df_clean_tca["clv"], 25) - 1.5 * iqr

mask_not_outlier = (df_clean_tca.clv <= upper_limit) & (df_clean_tca.clv >= lower_limit)
df_clean_tca_clv = df_clean_tca[mask_not_outlier]
display(df_clean_tca_clv)
sns.boxplot(df_clean_tca_clv['clv'])
```

```python
#eliminating outliars from mpa 
iqr = np.percentile(df_clean_tca_clv["mpa"], 75) - np.percentile(df_clean_tca_clv["mpa"], 25)
upper_limit = np.percentile(df_clean_tca_clv["mpa"], 75) + 1.5 * iqr
lower_limit = np.percentile(df_clean_tca_clv["mpa"], 25) - 1.5 * iqr

mask_not_outlier = (df_clean_tca_clv.mpa <= upper_limit) & (df_clean_tca_clv.mpa >= lower_limit)
df_clean = df_clean_tca_clv[mask_not_outlier]
display(df_clean)
sns.boxplot(df_clean['mpa'])
```

```python
#For the model:
 #Numerical features (the ones that are more significantly correlated to the target): number of policies, mpa and income
 #Categorical features: gender, marital status, employment, location, education    
```

## 2. Encoding Categorical Data

```python
#new df_model has only the data to be used in the model 

df_model = df_clean[["gender", "marital status", "es", "location code", "education", "income", "mpa", "number of policies", "tca", "clv"]]
df_model
```

```python
#rechecking correlations in df_model

annot = True
sns.heatmap(df_model.corr(method ='spearman'), annot=annot)
```

```python
#X are the explanatory features
#y is the target vector -> "customer lifetime value" (clv)



X = df_model.iloc[:, :-1]
y = df_model.iloc[:, -1]
```

```python
X.head(10)
```

```python
y.head(10)
```

```python
#Normalize (numerical) of X
#One hot encoding using pd dummies
X = pd.get_dummies(X, drop_first=True)
display(X)
print("Original df_model shape:",df_model.shape)
print("X Shape:",X.shape)
```

```python
# X shape has 7 new columns due to normalization (gender, marital status, es, location code and education)
```

## 3. Train-test split and Normalization

```python
# division of the dataset into trains and a test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
```

```python
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Modeling

## 1. Apply model

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression() 
```

```python
reg.fit(X_train, y_train)
```

```python
#coeficients of the regressor
reg.coef_
```

```python
predictions_train = reg.predict(X_train)
```

```python
predictions_train.size
```

```python
X_train.shape
```

# Model Validation

## 1. R2 (r squared)

```python
from sklearn.metrics import r2_score

predictions_test = reg.predict(X_test)

display(r2_score(y_train, predictions_train))
r2_score(y_test, predictions_test)
```
